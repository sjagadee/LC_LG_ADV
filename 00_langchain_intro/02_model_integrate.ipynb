## Multiple Model integrations
- OpenAI
- Google Gemini
- GROQ

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
```

### Open AI models

```python
model = init_chat_model(
    "openai:gpt-4o-mini",
    temperature=0.7,
)
model
```

```python
response = model.invoke("What is the current weather in Chicago?")
response
```

```python
response.content
```

#### Using ChatOpenAI

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model="gpt-4.1"
)
```

```python
response = model.invoke("Why do parrots speak?")
response
```

### Google gemini model

```python
model = init_chat_model(
    "google_genai:gemini-2.5-flash-lite",
    temperature=0.7,
)
model
```

```python
response = model.invoke("What is the current weather in Chicago?")
response
```

```python
response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite"
)
model
```

```python
response = model.invoke("Why do parrots speak?")
response
```

### GROQ model

```python
model = init_chat_model(
    "groq:qwen/qwen3-32b",
    temperature=0.7,
)
response = model.invoke("Why do parrots speak?")
response
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model= "qwen/qwen3-32b",
    temperature=0.7,
)
response = model.invoke("What is RAG?")
response
```
